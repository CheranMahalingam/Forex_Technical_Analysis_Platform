```python
import datetime
from datetime import datetime

def filled_dataframe(currency):
    time_array = pd.date_range(start="1/1/2018", freq="1H", periods=26304)
    ir_array = np.zeros(26304)
    ir_index = 0
    index = 0
    previous = 0
    for i in currency.iterrows():
        currency_datetime = datetime.strptime(i[1]["Time"], "%Y-%m")
        while index < 26304 and ir_index < 26304 and time_array[index] < currency_datetime:
            ir_array[ir_index] = previous
            ir_index += 1
            index += 1
        previous = i[1]["Interest Rate"]
    while ir_index < 26304:
        ir_array[ir_index] = currency["Interest Rate"].iloc[-1]
        ir_index += 1
    new_df = pd.DataFrame({"Time": time_array, "Interest Rate": ir_array})
    return new_df

filled_dataframe(aud_ir)
```

```python
aud_ir_datetime = filled_dataframe(aud_ir)
usd_ir_datetime = filled_dataframe(usd_ir)
cad_ir_datetime = filled_dataframe(cad_ir)
nzd_ir_datetime = filled_dataframe(nzd_ir)
eur_ir_datetime = filled_dataframe(eur_ir)
chf_ir_datetime = filled_dataframe(chf_ir)
jpn_ir_datetime = filled_dataframe(jpn_ir)
gbp_ir_datetime = filled_dataframe(gbp_ir)
gbp_ir_datetime
```
